# Activity: work with the iris dataset

1. Load the iris.csv file in your computer and understand the dataset


2. How many observations (rows) are in total?


3. How many variables (columns) are in total? What do they represent?


4. How many observations are for each type of flower?


5. What is the type of data for each variable?


6. What are the units of each variable?


```python
from sklearn.datasets import load_iris

# 1. Load dataset
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['species'] = iris.target
# Name columns
df['species'] = df['species'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})
```

```python

# Display the first 5 rows of the dataset
print("First 5 rows of the dataset:")
print(df.head())

# 2. How many observations (rows) are in total?
rows , cols = df.shape
print(f"\nTotal number of observations (rows): {rows}")

```

```python
# 3. How many variables (columns) are in total? What do they represent?
print(f"\nTotal number of variables (columns): {cols}")
print(f"\nColumn names and their representation:")

df.columns
```

```python
# 4. How many observations are for each type of flower?
flower_counts = df['species'].value_counts()
print(f"\nNumber of observations for each type of flower:")
print(flower_counts)

```

```python
# 5 and 6. What is the type of data for each variable and their units?
data_types = df.dtypes
print(f"\nData type for each variable:")
print(data_types)


```
