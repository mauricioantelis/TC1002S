```python
csv_path = "/content/cartwheel.csv"
df = pd.read_csv(csv_path)
display(df.head())
```

```python
print("Dimensiones del DataFrame:", df.shape)
print("Columnas del DataFrame:", df.columns.tolist())
print("Tipos de datos:")
print(df.dtypes)
```

```python
for col in df.select_dtypes(include=['object']).columns:
    print(f"Frecuencia de valores en {col}:")
    print(df[col].value_counts())
```

```python
print("Valores nulos por columna:")
print(df.isnull().sum())
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(df['Age'], kde=True, bins=10)
plt.title("Distribución de Age")
plt.show()

plt.figure(figsize=(12, 6))
sns.histplot(df['CWDistance'], kde=True, bins=10)
plt.title("Distribución de CWDistance")
plt.show()

plt.figure(figsize=(12, 6))
sns.histplot(df[['Age', 'Wingspan']], kde=True, bins=10)
plt.title("Distribución de Age y Wingspan")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(y=df['CWDistance'])
plt.title("Boxplot de CWDistance")
plt.show()

plt.figure(figsize=(8, 6))
sns.boxplot(y=df['Height'])
plt.title("Boxplot de Height")
plt.show()

plt.figure(figsize=(8, 6))
sns.boxplot(data=df[['CWDistance', 'Height', 'Wingspan']])
plt.title("Boxplots de CWDistance, Height y Wingspan")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Gender', y='CWDistance')
plt.title("Boxplot de CWDistance agrupado por Gender")
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Glasses', y='Height', hue='Gender')
plt.title("Boxplot de Height agrupado por Glasses y Gender")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Height', y='Wingspan')
plt.title("Scatter plot entre Height y Wingspan")
plt.show()

plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='CWDistance', y='Age', hue='Gender')
plt.title("Scatter plot entre CWDistance y Age, agrupado por Gender")
plt.show()
```

```python
sns.pairplot(df[['Age', 'CWDistance', 'Height', 'Wingspan']])
plt.show()
```
