```python
import pandas as pd
from sklearn import datasets

# Función para cargar dataset desde CSV o desde scikit-learn
def cargar_dataset(nombre_archivo, dataset_loader, columnas=None):
    try:
        df = pd.read_csv(nombre_archivo)
        print(f"Datos cargados exitosamente desde {nombre_archivo}")
    except FileNotFoundError:
        print(f"Archivo {nombre_archivo} no encontrado. Cargando el dataset desde scikit-learn.")
        dataset = dataset_loader()
        if columnas:
            df = pd.DataFrame(data=dataset.data, columns=columnas)
        else:
            df = pd.DataFrame(data=dataset.data)
        df['target'] = dataset.target
    print(f"El dataset tiene {df.shape[0]} filas y {df.shape[1]} columnas.")
    return df

# Función para analizar el dataset
def analizar_dataset(df, nombre):
    print(f"\nAnálisis del dataset {nombre}:")
    print(f"Número total de observaciones (filas): {df.shape[0]}")
    print(f"Número total de variables (columnas): {df.shape[1]}")
    print("\nTipos de datos por variable:")
    print(df.dtypes)

    if 'target' in df.columns:
        print("\nNúmero de observaciones por categoría:")
        print(df['target'].value_counts())

# Cargar, mostrar y analizar el dataset Iris
df_iris = cargar_dataset('iris.csv', datasets.load_iris, datasets.load_iris().feature_names)
print("Primeras filas del dataset Iris:")
print(df_iris.head())
analizar_dataset(df_iris, 'Iris')

# Cargar, mostrar y analizar el dataset Digits
df_digits = cargar_dataset('digits.csv', datasets.load_digits)
print("\nPrimeras filas del dataset Digits:")
print(df_digits.head())
analizar_dataset(df_digits, 'Digits')
```
