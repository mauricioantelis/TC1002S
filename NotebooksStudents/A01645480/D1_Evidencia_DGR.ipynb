## d) Elbow plot

Compute the Elbow plot

```python
inertias = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(X)
    inertias.append(kmeans.inertia_)

plt.plot(range(1, 11), inertias, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('Elbow Plot')
plt.show()
```

¿Cuál es el mejor número de clústeres K?

Según el gráfico del codo (elbow plot), el número óptimo de clústeres parece estar entre 3 y 4. En el gráfico, se observa una caída pronunciada en la inercia al aumentar de 1 a 3 clústeres, seguida de una disminución más gradual. El "codo" de la curva, que marca el punto donde la reducción de la inercia se vuelve menos significativa, suele considerarse el mejor número de clústeres. Aunque hay una ligera flexión alrededor del valor 3, la inercia sigue disminuyendo notablemente hasta el valor 4, lo que lo convierte en otra opción razonable dependiendo del equilibrio buscado entre simplicidad y reducción de inercia.

¿Este número de clústeres coincide con tu suposición inicial?

Mi suposición inicial fue de 3 clústeres, basada en una inspección visual de los gráficos de dispersión.
