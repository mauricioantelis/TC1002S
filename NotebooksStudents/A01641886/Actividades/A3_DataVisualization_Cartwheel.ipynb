# Frequency tables

The `value_counts()` method can be used to determine the number of times that each distinct value of a variable occurs in a data set.  In statistical terms, this is the "frequency distribution" of the variable. The `value_counts()` method produces a table with two columns.  The first column contains all distinct observed values for the variable.  The second column contains the number of times each of these values occurs.  Note that the table returned by `value_counts()` is actually a **Pandas** data frame, so can be further processed using any Pandas methods for working with data frames.


````python
# Number of times that each distinct value of a variable occurs in a data set
df.species.value_counts()
````

````python
# Proportion of each distinct value of a variable occurs in a data set
x = df.species.value_counts()
percentage = x/x.sum()
percentage
````

Note that the `value_counts()` method excludes missing values. We confirm this below by adding up observations to your data frame with some missing values and then computing `value_counts()` and comparing this to the total number of rows in the data set, which is 28. This tells us that there are 28 - (21+6) = 1 missing values for this variable (other variables may have different numbers of missing values).

````python
# Total number of observations
print('Numero de observaciones: ', df.shape[0])

# Total number of null observations
print('Numero de observaciones nulas: ', df.petal_width.isnull().sum())

# Total number of counts (excluding missing values)
print('Numero de observaciones validas: ', df.petal_width.notnull().sum())
````

# Histogram

It is often good to get a feel for the shape of the distribution of the data.

````python
# Plot histogram of petal length: Visualizes the distribution of petal length values.
sns.histplot(df.petal_length)
plt.show()

# Plot histogram of petal width: Displays the distribution of petal width values.
sns.histplot(df.petal_width)
plt.show()

# Plot histogram of sepal width: Shows the distribution of sepal width values.
sns.histplot(df.sepal_width)
plt.show()

# Plot histogram of sepal length: Illustrates the distribution of sepal length values.
sns.histplot(df.sepal_length)
plt.show()
````

````python
# Plot distribution of petal length with kernel density estimation: Visualizes the distribution of petal length values with a kernel density estimate.
sns.histplot(df.petal_length, kde=True)
plt.show()
````

````python
# Plot histogram of both sepal length and sepal width: Displays histograms for both sepal length and sepal width values.
df2plot = df[["sepal_length", "sepal_width"]]
sns.histplot(df2plot, kde=False)
plt.show()
````

# Histograms plotted by groups

While looking at a single variable is interesting, it is often useful to see how a variable changes in response to another. Thus, we can create a histograms of one quantitative variable grouped by another categorical variables.

````python
# Create FacetGrid for each species: Sets up a grid for multiple plots based on species.
g = sns.FacetGrid(df, row="species")

# Map histogram of petal width to the FacetGrid: Plots a histogram of petal width for each species using the FacetGrid.
g = g.map(plt.hist, "petal_width")
plt.show()

````

# Boxplots

Boxplots do not show the shape of the distribution, but they can give us a better idea about the center and spread of the distribution as well as any potential outliers that may exist. Boxplots and Histograms often complement each other and help an analyst get more information about the data

````python
# Create boxplot of sepal length: Generates a boxplot to visualize the distribution and spread of sepal length values.
sns.boxplot(y=df["sepal_length"])
plt.show()

# Create boxplot of sepal width: Illustrates the distribution and spread of sepal width values using a boxplot.
sns.boxplot(y=df["sepal_width"])
plt.show()

# Create boxplot of petal length: Displays the distribution and spread of petal length values using a boxplot.
sns.boxplot(y=df["petal_length"])
plt.show()

# Create boxplot of petal width: Visualizes the distribution and spread of petal width values using a boxplot.
sns.boxplot(y=df["petal_width"])
plt.show()
````

````python
# Create boxplots of petal width grouped by species: Generates boxplots of petal width for each species to compare their distributions and spreads.
sns.boxplot(data=df, x="petal_width", y="species")
plt.show()
````

````python
# Create boxplots of sepal width grouped by species: Illustrates the distribution and spread of sepal width values for each species using boxplots.
sns.boxplot(data=df, x="sepal_width", y="species")
plt.show()
````

# Scatter plot

Plot values of one variable versus another variable to see how they are correlated

````python
# scatter plot between two variables
sns.scatterplot(data=df, y="petal_width", x="species")
plt.show()
````

````python
# scatter plot between two variables (one categorical)
sns.scatterplot(data=df, y="sepal_width", x="species")
plt.show()
````

````python
# scatter plot between two variables (one categorical)
sns.scatterplot(data=df, y="sepal_length", x="species")
plt.show()
````

````python
# scatter plot between two variables grouped according to a categorical variable
sns.scatterplot(data=df, x="sepal_width", y="petal_width", hue="species")
plt.show()

````

````python
# scatter plot between two variables grouped according to a categorical variable
sns.scatterplot(data=df, x="sepal_length", y="petal_length", hue="species", size= "sepal_width")
plt.show()

````
