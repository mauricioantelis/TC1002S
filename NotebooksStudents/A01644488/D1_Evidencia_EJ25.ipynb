## d) Elbow plot

Compute the Elbow plot

```python
#Lista para guardar los errores
errorlist = []

k_values = range(1, 10) #Numero de clusters
for k in k_values:
    kmeanclust = KMeans(n_clusters=k, random_state=0, n_init="auto")
    kmeanclust.fit(X)
    errorlist.append(kmeanclust.inertia_)
    
sns.lineplot(x=k_values, y=errorlist, marker="o")
plt.title("Elbow plot")

```

```python
#Kmeans8row: Using only the 8 variables/pixels/features from the eighth row

km = KMeans(n_clusters=10, random_state=0, n_init="auto")

X  = dfdigits.iloc[:, 56:64].values

print(dfdigits.columns[56:64])

print (X.shape)

Cluster60 = km.fit_predict(X)

dfdigits["Cluster60"] = Cluster60

print(accuracy_score(digits.target, dfdigits["Cluster60"]))
print(km.inertia_)
```
