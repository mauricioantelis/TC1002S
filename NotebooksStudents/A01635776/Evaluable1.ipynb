Los comandos que he utilizado para poder llegar a el resultado anterior fueron los siguientes:

git clone: Crea una copia local de un repositorio remoto.

git add .: Agrega todos los cambios (archivos nuevos, modificados o eliminados). El punto (.) indica que se deben agregar todos los archivos del directorio actual.

git commit: Registra los cambios en el historial del repositorio. Después de agregar cambios con git add, debes hacer un commit para guardar esos cambios localmente. Normalmente se usa con un mensaje descriptivo (git commit -m "mensaje").

git pull: Descarga los cambios más recientes de un repositorio remoto y los integra con tu rama actual.

git push: Envía tus cambios locales a un repositorio remoto.
