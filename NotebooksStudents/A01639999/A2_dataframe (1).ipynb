```python
if "ColumnInserted" in df.columns:
    df.drop("ColumnInserted", axis=1, inplace=True)

df["AgeInMonths"] = df["Age"] * 12
display(df.head())

df.drop("AgeInMonths", axis=1, inplace=True)
display(df.head())

if "GenderGroup" in df.columns:
    df["GenderGroupNew"] = df.GenderGroup.replace({1: "Female", 2: "Male"})
    display(df.head())
    df.drop("GenderGroupNew", axis=1, inplace=True)
    display(df.head())

if "Height" in df.columns:
    df["ColumnStrata"] = pd.cut(df.Height, [60., 63., 66., 69., 72., 75., 78.])
    display(df.head())
    df.drop("ColumnStrata", axis=1, inplace=True)
    display(df.head())

unused_vars = ["ID", "GenderGroup", "GlassesGroup", "CompleteGroup"]
df.drop(columns=[col for col in unused_vars if col in df.columns], inplace=True)
display(df.head())
```

```python

unused_vars = ["ID", "GenderGroup", "GlassesGroup", "CompleteGroup"]
df.drop(columns=unused_vars, axis=1, inplace=True, errors='ignore')


selected_vars = ["Age", "Gender", "Glasses", "Height", "Wingspan", "CWDistance", "Complete", "Score"]
df = df[[col for col in selected_vars if col in df.columns]]

df.dropna(inplace=True)
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Definir los nombres de las columnas
colnames = ["Sepal_Length", "Sepal_Width", "Petal_Length", "Petal_Width", "Flower"]

# Cargar el dataset desde la URL proporcionada
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
df = pd.read_csv(url, header=None, names=colnames)

# Mostrar las primeras filas para verificar la correcta carga de datos
display(df.head())
```

```python
print(" Nombre de cada columna en el dataset:")
print(df.columns.tolist())

print("\n Tipo de datos de cada columna:")
print(df.dtypes)
```

```python
df_numeric = df.select_dtypes(include=['float64', 'int64'])

# Extraer medidas estadsticas manualmente
stats_summary = {
    "Min": df_numeric.min(),
    "Max": df_numeric.max(),
    "Mean": df_numeric.mean(),
    "Median": df_numeric.median(),
    "Std_Dev": df_numeric.std()
}

stats_summary_df = pd.DataFrame(stats_summary)

print("\n Estadsticas especficas (Min, Max, Mean, Median, Std Dev):")
display(stats_summary_df)
```

```python
print("\n Identificacin de valores nulos en cada columna:")
missing_values = df.isnull().sum()
print(missing_values)

# Si hay valores nulos, crear un nuevo dataset sin datos faltantes
if missing_values.sum() > 0:
    df_cleaned = df.dropna()
    print("\n Se ha creado un nuevo dataset sin valores nulos.")
    display(df_cleaned.head())
```

```python
# Dataset con solo Petal_Width, Petal_Length y Flower
df_petal = df[["Petal_Width", "Petal_Length", "Flower"]]
print("\n Dataset con Petal_Width y Petal_Length:")
display(df_petal.head())

# Dataset con solo Sepal_Width, Sepal_Length y Flower
df_sepal = df[["Sepal_Width", "Sepal_Length", "Flower"]]
print("\n Dataset con Sepal_Width y Sepal_Length:")
display(df_sepal.head())

# Dataset con Sepal_Width, Sepal_Length y Flower codificado numricamente
df_sepal_encoded = df_sepal.copy()
df_sepal_encoded["Flower"] = df_sepal_encoded["Flower"].astype("category").cat.codes  # Convierte las categoras en nmeros

print("\n Dataset con Sepal_Width, Sepal_Length y Flower codificado:")
display(df_sepal_encoded.head())
```

```python
df_petal.to_csv("iris_petal.csv", index=False)
df_sepal.to_csv("iris_sepal.csv", index=False)
df_sepal_encoded.to_csv("iris_sepal_encoded.csv", index=False)

print("\n Se han guardado los nuevos datasets como archivos CSV.")
```
