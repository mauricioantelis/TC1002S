# Kmeans clustering

Intuitive explanation

### Initialize/Preliminaries

```python
# Compute the number of points in the dataset
index = df.index  
Npoints = len(index)

print('The number of point is: ', Npoints)
```

## 1: Specify the number of clusters

Define the number $K$ of clusters

```python
# Let's assume our data has two clusters (note that the rest of the code is for K=2)
K  = 2
```

## 2: Initialize the centroids of the clusters

Randomly initialize the centroids of the clusters $C_1, C_2, \cdots, C_K$

```python
# Let's initialize the centroids for the K=2 clusters (this has to be done randomly)

# Let's do it manually
C1new =  np.array([1.5,3.0]) # 1st Centroid
C2new =  np.array([3.0,3.0]) # 2nd Centroid

# Print centroids

print("Centrode 1: ", C1new)
print("Centrode 2: ", C2new)

```

```python
# Plot the points and the centroids

plt.scatter(df.x1, df.x2, color = 'g',  s= 80)

plt.scatter(C1new[0], C1new[1], color = 'r', marker = 'o', label = 'Centrode 1', s=512, alpha = 0.8)
plt.scatter(C2new[0], C2new[1], color = 'b', marker = 'o', label = 'Centrode 1', s=512, alpha = 0.8)

plt.title("Data", fontsize=18)

plt.xlabel("X1")
plt.ylabel("X2")

plt.xlim(1,7)
plt.ylim(1,7)

plt.show()
```

## 3:  Repeat the following


Define the current centroids

```python
# Define centrode 1
C1 = C1new.copy() # 1st Centroid
C1
```

```python
# Define centrode 2
C2 = C2new.copy() # 2nd Centroid
C2
```

## 4: Assign each point  to its closest centroid

Compute the distance of each data point to each centroid

Assign each point to the centroid with the minimum Euclidean distance

Euclidean distance:
- Consider two points $x=(x_1,x_2)$ and  $y=(y_1,y_2)$.
- The Euclidean distance between these two points is
$ d = \sqrt {\left( {x_1 - x_2 } \right)^2 + \left( {y_1 - y_2 } \right)^2 } $


```python
# Compute the distance of each data point to each centroid

# Variable to save the distance of each point ot each centroid
Dis2Centroids = np.zeros((Npoints,K))

# Let's compute the distances manually
'''
ipoint, x1, x2 = 0, 2, 2
Dis2Centroids [ipoint,0] = np.sqrt((x1 - C1[0])**2 + (x2-C1[1])**2)
Dis2Centroids [ipoint,1] = np.sqrt((x1 - C2[0])**2 + (x2-C2[1])**2)

ipoint, x1, x2 = 1, 4, 3
Dis2Centroids [ipoint,0] = np.sqrt((x1 - C1[0])**2 + (x2-C1[1])**2)
Dis2Centroids [ipoint,1] = np.sqrt((x1 - C2[0])**2 + (x2-C2[1])**2)

ipoint, x1, x2 = 2, 2, 4
Dis2Centroids [ipoint,0] = np.sqrt((x1 - C1[0])**2 + (x2-C1[1])**2)
Dis2Centroids [ipoint,1] = np.sqrt((x1 - C2[0])**2 + (x2-C2[1])**2)

ipoint, x1, x2 = 3, 6, 6
Dis2Centroids [ipoint,0] = np.sqrt((x1 - C1[0])**2 + (x2-C1[1])**2)
Dis2Centroids [ipoint,1] = np.sqrt((x1 - C2[0])**2 + (x2-C2[1])**2)

ipoint, x1, x2 = 4, 6, 5
Dis2Centroids [ipoint,0] = np.sqrt((x1 - C1[0])**2 + (x2-C1[1])**2)
Dis2Centroids [ipoint,1] = np.sqrt((x1 - C2[0])**2 + (x2-C2[1])**2)

'''

for index, row in df.iterrows():
    Dis2Centroids [index,0] = np.sqrt((row['x1'] - C1[0])**2 + (row['x2']-C1[1])**2)
    Dis2Centroids [index,1] = np.sqrt((row['x1'] - C2[0])**2 + (row['x2']-C2[1])**2)


# Print results
print(Dis2Centroids)

```

```python
# Assign each point to the centroid with the minimum Euclidean distance

# Let's do this manually
#cluster = np.array([1,2,1,2,2])
#cluster = np.array([1,1,1,2,2])
#cluster = np.array([1,1,1,2,2])

cluster = np.array([np.argmin(point) + 1 for point in Dis2Centroids])

# Print results
print(cluster)
```

```python
# Scatter plot of the data asigning each point to the cluster it belongs to 

df1 = df[cluster==1]
df2 = df[cluster==2]

plt.scatter(df1.x1, df1.x2, s= 80, color = 'r', marker = 'o', label = 'Cluster 1') # 1 == rojo
plt.scatter(df2.x1, df2.x2, s= 80, color = 'b', marker = 's', label = 'Cluster 2') # 2 == azul


plt.scatter(C1[0], C1[1], color = 'r', marker = 'o', label = 'Centrode 1', s=512, alpha = 0.2) # 1 == rojo
plt.scatter(C2[0], C2[1], color = 'b', marker = 's', label = 'Centrode 1', s=512, alpha = 0.2) # 2 == azul

plt.title("Data", fontsize=18)

plt.xlabel("X1")
plt.ylabel("X2")

plt.xlim(1,7)
plt.ylim(1,7)

plt.show()
```

## 5: Compute the new centroid (mean) of each cluster

Compute the new centroid of each cluster

```python
# Let's compute the 1st Centroid

print(df1)

C1new[0] = df1.x1.mean()
C1new[1] = df1.x2.mean()

print(C1new)
```

```python
# Let's compute the 2nd Centroid

print(df2)

C2new[0] = df2.x1.mean()
C2new[1] = df2.x2.mean()

print(C2new)
```

```python
# Scatter plot of the data asigning each point to the cluster it belongs to 


df1 = df[cluster==1]
df2 = df[cluster==2]

plt.scatter(df1.x1, df1.x2, s= 80, color = 'r', marker = 'o', label = 'Cluster 1') # 1 == rojo
plt.scatter(df2.x1, df2.x2, s= 80, color = 'b', marker = 's', label = 'Cluster 2') # 2 == azul


plt.scatter(C1[0], C1[1], color = 'r', marker = 'o', label = 'Centrode 1', s=512, alpha = 0.2) # 1 == rojo
plt.scatter(C2[0], C2[1], color = 'b', marker = 's', label = 'Centrode 1', s=512, alpha = 0.2) # 2 == azul

plt.scatter(C1new[0], C1new[1], color = 'r', marker = 'o', label = 'Centrode 1', s=512, alpha = 0.8) # 1 == rojo
plt.scatter(C2new[0], C2new[1], color = 'b', marker = 's', label = 'Centrode 1', s=512, alpha = 0.8) # 2 == azul

plt.title("Data", fontsize=18)

plt.xlabel("X1")
plt.ylabel("X2")

plt.xlim(1,7)
plt.ylim(1,7)

plt.show()
```
