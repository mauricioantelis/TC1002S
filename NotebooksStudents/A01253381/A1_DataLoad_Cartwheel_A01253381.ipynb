```python
iris = pd.read_csv("iris.csv")
iris
```

```python
# How many observations (rows) are in total?
num_rows = len(iris.index)
print(f"Number of rows: {num_rows}")
```

```python
# How many variables (columns) are in total? What do they represent?
num_columns = len(iris.columns)
print(f"Number of columns: {num_columns}")
```

Hay 5 variables diferentes. Las columnas representan las medidas de los ptalos y spalos de 3 tipos de flores.

```python
# What is the type of data for each variable?
iris.info()
```

El tipo de dato de las medidas es float y class es un tipo objeto.
El atributo class es discreto y los atributos sepal_length,	sepal_width,	petal_length,	petal_width son contnuos.

```python
iris['class'] = iris['class'].astype('category')
iris.info()
```

```python
# How many observations are for each type of flower?
metrics = ['count']
iris.groupby(by='class').agg(metrics)
```
