## d) Elbow plot

Compute the Elbow plot

```python
# Intialize a list to hold sum of squared error (sse)
sse = []
# Define values of k
for k in range(1, 10):
# For each k
    km = KMeans(n_clusters=k, n_init="auto")
    km.fit(df[["x7","x11"]])
    sse.append(km.inertia_)

plt.style.use("fivethirtyeight")
plt.plot(range(1, 10), sse)
plt.xticks(range(1, 10))
plt.xlabel("Number of Clusters")
plt.ylabel("SSE")
plt.show()
```

## Do classification using the "digits" dataset

1) Load the dataset from "sklearn.datasets"

```python
from sklearn.datasets import load_digits
digits = load_digits()
```

2) Plot some of the observations  (add in the title the label/digit of that obserbation)

```python
plt.figure(figsize=(12, 6))
for index, (image, label) in enumerate(zip(digits.data[0:5], digits.target[0:5])):
    plt.subplot(1,5, index + 1)
    plt.imshow(np.reshape(image, (8,8)), cmap=plt.cm.gray)
    plt.title('Training: %i\n' % label, fontsize = 20)
```

3) Split the dataset in train and test

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(digits.data, digits.target, test_size=0.2, random_state=42)
```

4) Tune a classifier (Use the train set) in the following cases:

* ClassifierAll: Using all 64 variables/pixels/features

* Classifier1col: Using only the 8 variables/pixels/features from the firt column

* Classifier4col: Using only the 8 variables/pixels/features from the fourth column

* Classifier8col: Using only the 8 variables/pixels/
features from the eighth column

Note: in these four cases always use the same classification algorithm, e.g., a Suport Vector Machine


```python
from sklearn.svm import SVC
classifierall = SVC()
classifierall.fit(X_train, y_train)
classifier1col = SVC()
classifier1col.fit(X_train[:, :1], y_train)
classifier4col = SVC()
classifier4col.fit(X_train[:, 3:4], y_train)
classifier8col = SVC()
classifier8col.fit(X_train[:, 7:8], y_train)
```

5) Make predictions (use the test set)

```python
y_pred_all = classifierall.predict(X_test)
y_pred_1col = classifier1col.predict(X_test[:, :1])
y_pred_4col = classifier4col.predict(X_test[:, 3:4])
y_pred_8col = classifier8col.predict(X_test[:, 7:8])
```

6) Compute performance metrics

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred_all)
print(cm)
cm2 = confusion_matrix(y_test, y_pred_1col)
print(cm2)
cm3 = confusion_matrix(y_test, y_pred_4col)
print(cm3)
cm4 = confusion_matrix(y_test, y_pred_8col)
print(cm4)
```

7) Verify your results. Plot several observations from the same digit and add in the title the real label and the estimated label to check in what observations the classification was correct or incorrect

```python
digit=4
correct_indices = np.where((y_test == digit) & (y_pred_all == digit))[0]
incorrect_indices = np.where((y_test == digit) & (y_pred_all != digit))[0]
indices_to_plot = np.concatenate([correct_indices[:3], incorrect_indices[:3]])
plt.figure(figsize=(12, 6))
for index, i in enumerate(indices_to_plot):
    image = X_test[i]
    true_label = y_test[i]
    predicted_label = y_pred_all[i]

    plt.subplot(2, 3, index + 1)
    plt.imshow(np.reshape(image, (8, 8)), cmap=plt.cm.gray)
    plt.title(f"True: {true_label} Pred: {predicted_label}", fontsize=14)
    plt.axis("off")

plt.tight_layout()
plt.show()
```
