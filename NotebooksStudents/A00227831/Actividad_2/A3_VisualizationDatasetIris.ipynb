## Scatterplot

```python
# scatter plot between two variables
x = df.Ancho_Sepalo
y = df.Largo_Sepalo

plt.scatter(x, y, c ="blue")
plt.xlabel("Ancho")
plt.ylabel("Largo")
plt.title("Ancho x Largo Sepalo")
```

```python
# scatter plot between two variables (Categorico)

x = df.Ancho_Sepalo
y = df.Largo_Sepalo

plt.scatter(x, y, alpha = .6)
plt.xlabel("Ancho cm")
plt.ylabel("Largo cm")
plt.title("Ancho sepalo vs Largo sepalo")
```

```python
#Agrupado categorico
x = df.Ancho_Sepalo
y = df.Especie


sns.scatterplot(y,x, hue = df.Especie)
plt.xlabel("Tama√±o cm")
plt.ylabel("Especie")
plt.title("Ancho sepalo x especie")
```

```python
#Agrupado categorico
plt.subplot(1,1,1)
sns.scatterplot(x="Especie",y="Ancho_Sepalo",data=df)
plt.subplot(1,1,1)
sns.scatterplot(x="Especie",y="Largo_Sepalo",data=df)
plt.legend(["Ancho sepalo", "Largo Sepalo"])
plt.show()
```

```python
#Agrupado categorico
plt.subplot(1,1,1)
sns.scatterplot(x="Especie",y="Ancho_Sepalo",data=df, s=10, alpha = .5)
plt.subplot(1,1,1)
sns.scatterplot(x="Especie",y="Largo_Sepalo",data=df, s=60, alpha = .5)
plt.legend(["Ancho sepalo", "Largo Sepalo"])
plt.show()
```
