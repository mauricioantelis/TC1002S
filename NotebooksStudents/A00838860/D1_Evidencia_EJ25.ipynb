## d) Elbow plot

Compute the Elbow plot

```python

sse = []


k_rng = range(1,10)


for k in k_rng:
    
    km = KMeans(n_clusters=k, n_init="auto")
    
    km.fit_predict(df[['x10','x14']])
    
    sse.append(km.inertia_)

plt.plot(k_rng,sse)

plt.title('Elbow plot')
plt.xlabel('K')
plt.ylabel('Sum of squared error')
plt.show()
```

3) Do K means clustering in the following cases:

* KmeansAll: Using all 64 variables/pixels/features

* Kmeans1row: Using only the 8 variables/pixels/features from the firt row

* Kmeans4row: Using only the 8 variables/pixels/features from the fourth row

* Kmeans8row: Using only the 8 variables/pixels/
features from the eighth row


```python

kmeansAll = KMeans(n_clusters=10, n_init="auto", random_state=42)
Cluster1 = kmeansAll.fit_predict(df.drop("target", axis=1) )

kmeans1row = KMeans(n_clusters=10, n_init="auto", random_state=42)
Cluster2 = kmeans1row.fit_predict(df.iloc[:,:8])

kmeans4row = KMeans(n_clusters=10, n_init="auto", random_state=42)
Cluster3 = kmeans4row.fit_predict(df.iloc[:,24:32])

kmeans8row = KMeans(n_clusters=10, n_init="auto", random_state=42)
Cluster4 = kmeans8row.fit_predict(df.iloc[:,56:64])

df['Cluster1'] = Cluster1
df['Cluster2'] = Cluster2
df['Cluster3'] = Cluster3
df['Cluster4'] = Cluster4

#Ajustando los clusters para que coincidan con su digito
cluster_digit_count = pd.crosstab(df['Cluster1'], df['target'])
cluster_to_digit = cluster_digit_count.idxmax(axis=1)
df["Cluster1"]=df["Cluster1"].map(cluster_to_digit)

cluster_digit_count = pd.crosstab(df['Cluster2'], df['target'])
cluster_to_digit = cluster_digit_count.idxmax(axis=1)
df["Cluster2"]=df["Cluster2"].map(cluster_to_digit)

cluster_digit_count = pd.crosstab(df['Cluster3'], df['target'])
cluster_to_digit = cluster_digit_count.idxmax(axis=1)
df["Cluster3"]=df["Cluster3"].map(cluster_to_digit)

cluster_digit_count = pd.crosstab(df['Cluster4'], df['target'])
cluster_to_digit = cluster_digit_count.idxmax(axis=1)
df["Cluster4"]=df["Cluster4"].map(cluster_to_digit)
```

5) Compute the Elbow plot

```python
sse = []


k_rng = range(1,30)

# For each k
for k in k_rng:
    # Create model
    km = KMeans(n_clusters=k, n_init="auto")
    # Do K-means clustering
    km.fit_predict(df.drop("target", axis=1))
    # Save sse for each k
    sse.append(km.inertia_)

# Plot sse versus k
plt.plot(k_rng,sse)

plt.title('Elbow plot')
plt.xlabel('K')
plt.ylabel('Sum of squared error')
plt.show()
```

* ClassifierAll: Using all 64 variables/pixels/features

Note: in these four cases always use the same classification algorithm, e.g., a Suport Vector Machine


```python
from sklearn.ensemble import RandomForestClassifier


rf_all = RandomForestClassifier(n_estimators=100, max_depth=None, min_samples_split=2, 
                                min_samples_leaf=1, max_features="sqrt", random_state=42)

rf_all.fit(X_train, y_train)



X_train_1col = X_train.iloc[:, :8]
X_test_1col = X_test.iloc[:, :8]


rf_1col = RandomForestClassifier(n_estimators=100, max_depth=None, min_samples_split=2, 
                                 min_samples_leaf=1, max_features="sqrt", random_state=42)


rf_1col.fit(X_train_1col, y_train)




X_train_4col = X_train.iloc[:, 24:32]
X_test_4col = X_test.iloc[:, 24:32]


rf_4col = RandomForestClassifier(n_estimators=100, max_depth=None, min_samples_split=2, 
                                 min_samples_leaf=1, max_features="sqrt", random_state=42)


rf_4col.fit(X_train_4col, y_train)




X_train_8col = X_train.iloc[:, 56:64]
X_test_8col = X_test.iloc[:, 56:64]


rf_8col = RandomForestClassifier(n_estimators=100, max_depth=None, min_samples_split=2, 
                                 min_samples_leaf=1, max_features="sqrt", random_state=42)


rf_8col.fit(X_train_8col, y_train)
```

```python
pred_all = rf_all.predict(X_test)
pred_1col = rf_1col.predict(X_test.iloc[:, 0:8]) 
pred_4col = rf_4col.predict(X_test.iloc[:, 24:32])  
pred_8col = rf_8col.predict(X_test.iloc[:, 56:64])

print("Valores reales")
print(y_test[:15].values)
print("Predicciones clasificador con todas las variables")
print(pred_all[:15])
print("Predicciones clasificador con solo la primera columna")
print(pred_1col[:15])
print("Predicciones clasificador con solo la cuarta columna")
print(pred_4col[:15])
print("Predicciones clasificador con solo la octava columna")
print(pred_8col[:15])
```
