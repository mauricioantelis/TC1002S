```python
#3.Crear una nueva dataset que contenga solo petal width y length and the type of Flower
df1 = df[["PetalLength", "PetalWidth", "Class"]]
df1.to_csv('myDataFrameW-LPetal_Iris.csv')
df1
```

```python
#5.Crear nuevo dataset que contenga setal width y length y el tipo de flor como columna numerica
df["ClassNum"] = df.Class.replace({"Iris-setosa": 1, "Iris-versicolor": 2, "Iris-virginica": 3})
df.to_csv('myDataFrameClassNum_Iris.csv')
df
```
