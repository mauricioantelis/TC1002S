4. How many observatoins are for each type of flower? 

```python
counter = 0

for i in range(0,Nr):
    curr = df[4][i]
    nex = df[4][i+1] if (i != Nr-1) else "s"
    if (curr != nex):
        print (df[4][i], "contains ",counter+1, " observations on DF")
        counter = 0
    else:
        counter += 1
```
