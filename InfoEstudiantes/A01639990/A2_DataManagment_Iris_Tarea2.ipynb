# Add and eliminate columns

```python
df.head()
```

```python
# Create a column data
NewColumnData = df.Petal_length_in_cm*10

# Insert that column in the data frame
df.insert(5, "Petal_length_in_mm", NewColumnData, True)

df.head()
```

```python
#Eliminate inserted column
df.drop("Petal_length_in_mm", axis=1, inplace = True)
```

```python
df.head()
```

```python
# # The new column is a function of another column
df["Sepal_length_in_mm"] = df["Sepal_length_in_cm"] * 10
# 
df.head()
```

```python
# # Eliminate inserted column
df.drop("Sepal_length_in_mm", axis=1, inplace = True)
#
df.head()
```

```python
# Add a new column with text labels reflecting the code's meaning

df["ClassNumber"] = df.Class.replace({"Iris-setosa":1, "Iris-versicolor":2, "Iris-virginica":3})

# Show the first 5 rows of the created data frame
df.head()
```

```python
## Eliminate inserted column
df.drop("ClassNumber", axis=1, inplace = True)
##df.drop(['GenderGroupNew'],vaxis='columns',vinplace=True)

df.head()
```

```python
## Add a new column with strata based on these cut points
#
## Create a column data
NewColumnData = df.Sepal_length_in_cm/df.Sepal_length_in_cm
#
## Insert that column in the data frame
df.insert(1, "ColumnStrata", NewColumnData, True)
#
df["ColumnStrata"] = pd.cut(df.Petal_length_in_cm, [1., 2., 3., 4., 5., 6., 7.])
#
## Show the first 5 rows of the created data frame
df.head()
```

```python
## Eliminate inserted column
df.drop("ColumnStrata", axis=1, inplace = True)
#
df.head()
```

```python
# Drop several "unused" columns
vars = ["Sepal_length_in_cm","Sepal_width_in_cm"]
df.drop(vars, axis=1, inplace = True)

df.head()
```
